取得 Name 和 Salary 二欄資料

```python
import pandas as pd

data = {
    'ID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace', 'Helen', 'Ivy', 'Jack'],
    'Age': [25, 30, 35, 40, 22, 29, 38, 27, 33, 31],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix', 'Philadelphia', 'San Antonio', 'San Diego', 'Dallas', 'San Jose'],
    'Salary': [50000, 60000, 75000, 80000, 45000, 62000, 72000, 58000, 69000, 71000],
    'Department': ['IT', 'HR', 'Finance', 'Marketing', 'IT', 'HR', 'Finance', 'Marketing', 'IT', 'Finance']
}
df = pd.DataFrame(data)
df
```

將薪水高於第3四分位數的人員標示為高薪

```python
hs = df['Salary'].quantile(0.75) #四分位數第三位
df['HighSalary'] = df['Salary'].apply(lambda x: True if x > hs else False)
df
```

```python
df[df['HighSalary']] #取出高薪者
```

刪除高薪欄位

```python
df = df.drop(columns=['HighSalary'])
df
```

用pd.concat() 新增資料（資料集）

```python
new_data = [
    {'ID': 111, 'Name': 'Kevin', 'Age': 28, 'City': 'Austin', 'Salary': None, 'Department': 'IT'},
    {'ID': 112, 'Name': 'Lily', 'Age': 26, 'City': 'Seattle', 'Salary': 55000, 'Department': 'Marketing'},
    {'ID': 112, 'Name': 'Lily', 'Age': 26, 'City': 'Seattle', 'Salary': 55000, 'Department': 'Marketing'}
]
new_df = pd.DataFrame(new_data)
df = pd.concat([df, new_df], ignore_index=True)
df
```

```python
corr_matrix = df[['Age', 'Salary']].corr() #年齡、薪水相關係數
corr_matrix
```

找出空值的資料列，並填充空值（薪水依該部門平均值）

```python
# 計算每一欄有多少個 NaN
df.isnull().sum()
```

```python
avg_dp = df.groupby('Department')['Salary'].mean() #群組化部門與平均薪資
avg_dp
```

```python
for i in range (len(df)):
  if pd.isnull(df.loc[i, 'Salary']):
    df.loc[i, 'Salary']= avg_dp[df.loc[i, 'Department']]
df
```

刪除重複資料

```python
df = df.drop_duplicates() #刪除重複資料
```

新增欄位 'NewSalary'，如果薪水小於平均值則薪水加 500，否則維持原薪水；新增欄位 'RankSalary' 填入薪水排名。

```python
avg_salary = df['Salary'].mean() #薪水平均值
avg_salary
```

```python
df['NewSalary'] = df['Salary'].apply(lambda x:x +500 if x < avg_salary else x) #如果薪水小於平均值則薪水加 500，否則維持原薪水；新增欄位 'RankSalary' 填入薪水排名。
df
```

```python
df['RankSalary'] = df['Salary'].rank(ascending=False) #薪水排名
df
```

```python
df[['Salary','NewSalary','RankSalary']] = df[['Salary','NewSalary','RankSalary']].astype(int) #轉換資料型態為整數
df
```
